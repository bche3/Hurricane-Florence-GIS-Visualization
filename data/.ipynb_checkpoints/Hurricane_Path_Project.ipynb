# Plotting Path of Hurricane Florence using geospatial data

Geographic Information Systems (GIS) allows us to integrate technology of geospatial data to analyze and visualize the path for Hurricane Florence from the dates of August 30th to September 18th in 2018


The Data
- http://flhurricane.com/cyclone/stormhistory.php?storm=6&year=2018
- https://eric.clst.org/tech/usgeojson/

```python
 # import packages
import geopandas
import numpy as np
import pandas as pd
from shapely.geometry import Point

import missingno as msn

import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# Check the type of dataframe that was loaded in
type(country)
```

```python
# Check the data type of the column containing coordinates
type(country.geometry) 
```

Each value in the Geoseries is a Shapely object. Each object can be used for a different type of physical object such as:
-  Point: for building
- Line: for Street
- Polygon: for city
- MultiPolygon: for country with multiple cities inside

```python
florence = pd.read_csv("/Users/brian/Documents/GitHub/GIS-Hurricane-Path-Project/data/florence_storm_history.csv")
florence.head()
```

# Exploratory Data Analysis

```python
florence.info()
```

```python
# Notice you can always adjust the color of the visualization
msn.bar(florence, color='darkolivegreen');
```

```python
# Statistical information
florence.describe()
```

```python
# dropping all unused features:
florence = florence.drop(['AdvisoryNumber', 'Forecaster', 'Received'], axis=1)
florence.head()
```

```python
# Add "-" in front of the number to correctly plot the data:
florence['Long'] = 0 - florence['Long']
florence.head()
```

```python
# Combining Lattitude and Longitude to create hurricane coordinates:
florence['coordinates'] = florence[['Long', 'Lat']].values.tolist()
florence.head()
```

```python
# Change the coordinates to a geoPoint
florence['coordinates'] = florence['coordinates'].apply(Point)
florence.head()
```

```python
print(type(florence))
print(type(florence['coordinates']))
```

```python
# Convert the count df to geodf
florence = geopandas.GeoDataFrame(florence, geometry='coordinates')
florence.head()
```

```python
print(type(florence))
print(type(florence['coordinates']))
```

```python
# Filtering from before the hurricane was named.
florence[florence['Name']=='Six']
```

```python
florence.groupby('Name').Type.count()
```

```python
print("Mean wind speed of Hurricane Florence is {} mph and it can go up to {} mph maximum".format(round(florence.Wind.mean(),4),
                                                                                         florence.Wind.max()))
```

# Visualization

```python
florence.plot(figsize=(20,10));
```

```python
 # Plotting to see the hurricane overlay the US map:
fig, ax = plt.subplots(1, figsize=(30,20))
base = country[country['NAME'].isin(['Alaska','Hawaii']) == False].plot(ax=ax, color='#3B3C6E')

# plotting the hurricane position on top with red color to stand out:
florence.plot(ax=base, color='darkred', marker="*", markersize=10);
```

```python
 fig, ax = plt.subplots(1, figsize=(20,20))
base = country[country['NAME'].isin(['Alaska','Hawaii']) == False].plot(ax=ax, color='#3B3C6E')
florence.plot(ax=base, column='Wind', marker="<", markersize=10, cmap='cool', label="Wind speed(mph)")
_ = ax.axis('off')
plt.legend()
ax.set_title("Hurricane Florence in US Map", fontsize=25)
plt.savefig('Hurricane_footage.png',bbox_inches='tight');
```

The hurricane was strongest when it's offshore near the east coast. As it approached the land, the hurricane started losing its strength, but at the wind speed in range 60 to 77 miles per hour, it can still make horrible damages.
